first, we gotta read the data in

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('dat.tsv', sep=',')
df.head(3)
```

the suitable graph for this is the **scatter plot**. this is because we want to have one axis with one variable and another axis with a diferent variable (as goes like the first sentence or two of 12.3.4). also, the next question is asking about if they have a relationship and a scatter plot would help determine if they do.

```python
fig,ax = plt.subplots(dpi=300, figsize=(10,8))

ax.scatter(df['T'], df['k'])
ax.set(xlabel='Temperature ($^\circ$C)', ylabel='k (W/m$^\circ$C)')
```

now calculating our $r$ (using the formula from ch16)

```python
x = df['T']
y = df['k']

xbar,ybar = x.mean(), y.mean()
numerator = ((x-xbar) * (y-ybar)).sum()
denominator = np.sqrt(((x-xbar)**2).sum())  *  np.sqrt(((y-ybar)**2).sum())
r = numerator / denominator
print(f'r: {r:.4f}')
```

the sign (positive) suggests a positive correlation i.e. as temperature increases so does k. the magnitude says that this relationship is quite strong and fairly linear.
